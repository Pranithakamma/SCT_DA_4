```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
df = pd.read_csv("marketing_campaign.csv")
df.head()
```

```python
#Basic Data Exploration
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
#Data Cleaning
print("Original Shape:", df.shape)
df = df.dropna()
df = df.drop_duplicates()

print("Cleaned Shape:", df.shape)
df.head()
```

```python
#Analyze Marketing Channels
#Average ROI by Channel
channel_roi = df.groupby("Channel_Used")["ROI"].mean().sort_values(ascending=False)
print(channel_roi)

#Total Clicks by Channel
channel_clicks = df.groupby("Channel_Used")["Clicks"].sum().sort_values(ascending=False)
print(channel_clicks)

#Total Impressions by Channel
channel_impressions = df.groupby("Channel_Used")["Impressions"].sum().sort_values(ascending=False)
print(channel_impressions)

```

```python
#Visualization (Correct),#ROI by Channel
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
channel_roi.plot(kind="bar")
plt.title("Average ROI by Marketing Channel")
plt.xlabel("Channel")
plt.ylabel("ROI")
plt.show()
```

```python
#Clicks by Channel
plt.figure(figsize=(8,5))
channel_clicks.plot(kind="bar")
plt.title("Total Clicks by Marketing Channel")
plt.xlabel("Channel")
plt.ylabel("Clicks")
plt.show()
```

```python
#Impressions by Channel
plt.figure(figsize=(8,5))
channel_impressions.plot(kind="bar")
plt.title("Total Impressions by Marketing Channel")
plt.xlabel("Channel")
plt.ylabel("Impressions")
plt.show()
```

```python
#Relationship Analysis,Clicks vs Impressions
import seaborn as sns
plt.figure(figsize=(7,5))
sns.scatterplot(x="Impressions", y="Clicks", data=df)
plt.title("Impressions vs Clicks")
plt.show()
```

```python
#Engagement Score by Channel
plt.figure(figsize=(8,5))
sns.barplot(x="Channel_Used", y="Engagement_Score", data=df)
plt.title("Engagement Score by Channel")
plt.show()
```

# Business Insights Report

After analyzing the marketing campaign dataset, several key patterns were observed.

Marketing channels differ significantly in terms of performance. Some channels generate higher ROI and engagement compared to others. Channels with higher impressions generally produce more clicks, indicating strong audience reach.

However, higher reach does not always guarantee better ROI, showing the importance of efficient marketing strategies.

Customer segments also respond differently to campaigns, suggesting the need for targeted marketing.

## Recommendations:
- Focus more budget on high ROI channels
- Optimize low-performing channels
- Target specific customer segments for better results
- Improve campaign strategies to increase engagement and conversions

## Conclusion:
Data-driven decision-making can significantly improve marketing performance and return on investment.
